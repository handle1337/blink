# Set-up

Before commencing, make sure to retrieve a sample from CasJobs by using the following query under the context of the "HLSP_ATLAS_VAR" database: 
```sql
SELECT TOP 10000 ATO_ID,ra,dec into mydb.variables from object
```

````python
import io
import re
import sys
import time
import requests


import configparser as cp
import pandas as pd


config = cp.ConfigParser(interpolation=None)


config.read("../config.ini")

username = config['LOGIN']['username']
password = config['LOGIN']['password']

if (not username) or (not password):

    username = input("Enter your username: ")
    password = input("Enter your password: ")

print(f'username: {username} pass: {password}')

BASEURL = "https://fallingstar-data.com/forcedphot"



resp = requests.post(url=f"{BASEURL}/api-token-auth/", data={'username': username, 'password': password})

if resp.status_code == 200:
    token = resp.json()['token']
    print(f'Your token is {token}')
    headers = {'Authorization': f'Token {token}', 'Accept': 'application/json'}
else:
    print(f'ERROR {resp.status_code}')
    print(resp.json())
````

## Fetching forced photometry

We first use the CSV file provided by CasJobs using our previous SQL query to obtain the RA and Dec of our variable star dataset.

We have included a sample file `variables.csv` in this project as well containing some data, replace this file under the projects path with your desired CSV.


````python
import csv

with open('variables.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        id = row[0]
        ra = row[1]
        dec = row[2]
        print(id + ra+ dec)
````
